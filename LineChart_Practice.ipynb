```python
import matplotlib.pyplot as plt
```

```python
# data to Plot 
samosa =   [56,87,90,95,120,34]
days = ["Mon","Tue","Wed","Thurs","Fri","Sat"]

# color marker line 
# Shorhand on formatting

plt.plot(days,samosa,'go:')

# To add Title 
plt.title("Number of Samosa sold per day", fontsize=14)

# To add Labels in both x and y axis  

plt.xlabel("Days",fontsize=12)
plt.ylabel("No.Samosa sold",fontsize=12)

# To add Legend
plt.legend(['Samosa Sales'])

# To add Text 
plt.text('Tue',50,'Sales of Samosa Per Day',fontsize=12, alpha=1,color="blue",rotation=20)

plt.show

# help(plt.plot)
```

```python
years = range(2010, 2022)
maths= [89, 91, 91, 92, 92, 93, 93, 96, 97, 95, 92, 93]
science= [96, 94, 93, 92, 91, 90, 97, 90, 91,  89, 90, 89 ]
```

```python
# Plot Maths marks 
plt.plot(years, maths)

# plot Science marks
plt.plot(years, science)

# Adding Labels  
plt.xlabel('Year')
plt.ylabel('Marks')
```

```python
plt.plot (years,maths)
plt.plot (years, science)

plt.xlabel('Year')
plt.ylabel('Maths Vs Science')

plt.title("Marks")
plt.legend(['Maths', 'Science'])
```

```python
plt.plot(years, maths, marker='s', c='b', ls='-', lw=2, ms=8, mew=1, mec='white')
#ms - marker size
#mew - marker edge width
#mec - marker edge color
plt.plot(years, science, marker='o', c='r', ls='--', lw=3, ms=10, alpha=.2)

plt.xlabel('Year')
plt.ylabel('Maths Vs Science')

plt.title("Marks")
plt.legend(['maths', 'science']);

# To add grid
plt.grid(color='#95a5a6', linestyle='--', linewidth=2, axis='y', alpha=0.7)
```

```python
# To change the size of the figure
plt.figure(figsize=(8, 8))
plt.plot(years,science)
plt.plot(years,maths)
```

```python
df=pd.read_csv('sciencenmaths.csv')
df
```

```python
# list the columns 
df.columns
```

```python
# remove space from the column name 
df.columns = df.columns.str.replace(' ','')
```

```python
years = df['year'] # or years = range(2017,2022)
maths = df['Maths']
science = df['Science']
english = df['English']
```

```python
plt.figure(figsize=(10,6))
plt.plot(years,maths)
plt.plot(years,science)
plt.plot(years,english)
plt.legend(['Maths','Science','English'])
plt.xlabel('Year',fontsize=12)
plt.ylabel('Marks',fontsize=12)
plt.title('Summarized marks')
plt.plot()
```
