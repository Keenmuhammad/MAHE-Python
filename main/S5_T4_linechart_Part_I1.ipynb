Line Chart 
 
        - Series of data point connected by staright line. 


Appication 

       - Used to track changes over the short or long period of time. 
       - Used to compare the changes of two different values over the same period of time.

Example

      - Analyse how students performance changed over the period of time.
      - Compare marks between two different subejcts.

Limitation

          - Mess the chart if many categories are compared in one line chart

```python
#import packages 

import matplotlib.pyplot as plt
```

```python
# data to Plot 
samosa =   [56,87,90,95,120,34]
days = ["Mon","Tue","Wed","Thurs","Fri","Sat"]

# color marker line 
# Shorhand on formatting

plt.plot(days,samosa,'go:')

# To add Title 
plt.title("Number of Samosa sold per day", fontsize=14)

# To add Labels in both x and y axis  

plt.xlabel("Days",fontsize=12)
plt.ylabel("No.Samosa sold",fontsize=12)

# To add Legend
plt.legend(['Samosa Sales'])

# To add Text 
plt.text('Tue',50,'Sales of Samosa Per Day',fontsize=12, alpha=1,color="red",rotation=20)

plt.show

# help(plt.plot)
```

# Plotting Multiple Axis

```python
years = range(2010, 2022)
maths= [89, 91, 91, 92, 92, 93, 93, 96, 97, 95, 92, 93]
science= [96, 94, 93, 92, 91, 90, 97, 90, 91,  89, 90, 89 ]
```

```python
# Plot Maths marks 
plt.plot(years, maths)

# plot Science marks
plt.plot(years, science)

# Adding Labels  
plt.xlabel('Year')
plt.ylabel('Marks')
```

# Adding Legend

```python
plt.plot (years,maths)
plt.plot (years, science)

plt.xlabel('Year')
plt.ylabel('Maths Vs Science')

plt.title("Marks")
plt.legend(['Maths', 'Science'])
```

# More on Formatting

```python
plt.plot(years, maths, marker='s', c='b', ls='-', lw=2, ms=8, mew=1, mec='white')
#ms - marker size
#mew - marker edge width
#mec - marker edge color
plt.plot(years, science, marker='o', c='r', ls='--', lw=3, ms=10, alpha=.2)

plt.xlabel('Year')
plt.ylabel('Maths Vs Science')

plt.title("Marks")
plt.legend(['maths', 'science']);

# To add grid
plt.grid(color='#95a5a6', linestyle='--', linewidth=2, axis='y', alpha=0.7)
```

# Changing figure size

```python
# To change the size of the figure
plt.figure(figsize=(12, 12))
plt.plot(years,science)
plt.plot(years,maths)
```

Styling Lines and Markers

The plt.plot function supports many arguments for styling lines and markers:

    color or c: Set the color of the line (supported colors)
    linestyle or ls: Choose between a solid or dashed line
    linewidth or lw: Set the width of a line
    markersize or ms: Set the size of markers
    markeredgecolor or mec: Set the edge color for markers
    markeredgewidth or mew: Set the edge width for markers
    markerfacecolor or mfc: Set the fill color for markers
    alpha: Opacity of the plot
