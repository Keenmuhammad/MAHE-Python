Cleveland Dot Plot 

      - created by William S.Cleveland.
      - Use position to plot the value.
      - similar to bar graph.

Wikinson Dot Plot 

    - Created by Leland Wikinson.
    - Displays a distribution of continuous data points.
    - Similar to histogram, but it display individual data points instead of bins.

Application 

        - To display the distribution of data 
        - To cluster the value 
        - To find outliers

```python


# pyplot is a submodule comes under matplotlib
y1= [4,7,3,6,1,8,9,2,4,7,4,3]
y2= [7,7,8,10,15,6,4,8,8,12,11,11]
x= ['Sl','S2', 'S3','S4','S5','S6','S7','S8','S9','S10','S11','S12']
plt.plot(x, y1 , 'o')

plt.plot(x,y2,'r*')

plt.title("Spending Time in Social Media - WeekDay Vs. WeekEnd")
plt.legend(['WeekDay','WeekEnd'])

# axis labeling
plt.xlabel('Student')
plt.ylabel('No.of Hours')

# adding grid 
plt.grid(alpha=0.4, linestyle='--',color='#0000')

```

```python
data  = [4,7,3,6,1,8,9,2,4,7,4,3]
pos = [] 
keys = {} # this dict will help to keep track ...

# this loop will give us a list of frequencies to each number
for num in data: 
   if num not in keys:
      keys[num] = 1
      pos.append(1)
   else:
      keys[num] += 1
      pos.append(keys[num])


print(pos)

plt.plot(data, pos,'o')
plt.xlabel('No. of Hours')
plt.ylabel('No. of Students')
plt.show()
```
