Naming convention 

    A variable name 
            -  must start with a letter or the underscore character.
            -  can only contain alpha-numeric characters and underscores    (A-z, 0-9, and _ ).
            -  case-sensitive (course, Course and COURSE are three different variables).

            - cannot start with numeric value. 
            - reserved words(keywords) cannot be used naming the variable.
            - Do not use white space between two words

```python
# declare variable 
# Basic syntax : <variable_name> = value 
fruit ="Apple"
```

```python
# Case-sensitive
print(fruit)
```
