```python
# import Packages 
import matplotlib.pyplot as plt
```

```python
sizes = [35, 34, 25, 18]
labels = 'Machine Learning', 'Embedded', 'VLSI', 'Information Science'

plt.pie(sizes, labels = labels)

plt.title('Breakdown of Student Strength by Course')

plt.show()
```

```python
sizes = [35, 34, 25, 18]
labels = 'Machine Learning', 'Embedded', 'VLSI', 'Information Science'

# Plot values using Pie Chart
# autopct enables you to display the percent value

plt.pie(sizes, labels = labels, autopct='%1.2f%%')


# To add Title
plt.title('Breakdown of Student Strength by Course')

plt.show()
```

```python
sizes = [35, 34, 25, 18]
labels = 'Machine Learning', 'Embedded', 'VLSI', 'Information Science'

# To change colors 
plt.pie(sizes, labels = labels, autopct='%1.2f%%',colors=['violet','lightgreen','tomato','r'])

plt.title('Breakdown of Student Strength by Course')

plt.show()
```

```python
# Explode: To emphasis the slices 
sizes = [35, 34, 25, 18]
labels = 'Machine Learning', 'Embedded', 'VLSI', 'Information Science'

plt.pie(sizes, labels = labels, autopct='%1.2f%%',colors=['violet','lightgreen','tomato','r'],explode=(0.2,0.1,0,0))

plt.title('Breakdown of Student Strength by Course')

plt.show()
```

```python
sizes = [35, 34, 25, 18]
labels = 'Machine Learning', 'Embedded', 'VLSI', 'Information Science'

# To change the figure size
plt.figure(figsize=(8,6))

# to change the start angle 
plt.pie(sizes, labels = labels, autopct='%.2f%%',colors=['violet','lightgreen','tomato','m'],explode=(0.2,0,0,0), shadow=True, startangle=90)

plt.title('Breakdown of Student Strength by Course')

# To add Legend
plt.legend(labels, loc='lower right',title="course name:")
plt.show()
```

```python
# To create Circle 
circle = plt.Circle((0, 0), 0.4, color='w')


d = plt.pie(sizes, labels=labels, autopct='%.1f%%',
            startangle=90, labeldistance=1)

# To add Circle in Pie chart
plt.gca().add_artist(circle)
            
plt.show()
```
